```python
import pandas as pd

listings = pd.read_csv(
    "Listings.csv",
    encoding="ISO-8859-1",
    low_memory=False
    )
listings.head()
```

```python
listings.info()
```

```python
listings["host_since"] = pd.to_datetime(listings["host_since"])

listings.info()
```

```python
paris_listings = (
    listings.query("city == 'Paris'")
    .loc [:,["host_since", "neighbourhood", "city", "accommodates", "price"]]
)
paris_listings.info()
```

```python
paris_listings.isna().sum()
```
